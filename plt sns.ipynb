# #grids

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
a=sns.PairGrid(iris)
a.map(plt.scatter)
```

```python
a=sns.PairGrid(iris)
a.map_diag(sns.distplot)
a.map_upper(plt.scatter)
a.map_lower(sns.kdeplot)
```

```python
b=sns.FacetGrid(data=tips,row='smoker',col='time')
b.map(sns.distplot,'total_bill')
```
